```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
##Divide the dataset into Dependent and Independent Feautures
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
X.head()
```

```python
y.head()
```

```python
##Feautrue Engineering
geo = pd.get_dummies(X['Geography'],drop_first=True)
gender = pd.get_dummies(X['Gender'],drop_first= True)
```

```python
geo
```

```python
gender
```

```python
X = X.drop(['Geography','Gender'],axis = 1)
```

```python
pd.concat([X,geo,gender],axis = 1)
```

```python
#Splitting the dataset into training and test dataset 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

sc = StandardScaler()  

X_train = sc.fit_transform(X_train)  
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
## part 2 Now let's create ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
##intialize ANN
classifier = Sequential()
```

```python
#Adding Input layer
classifier.add(Dense(units=11,activation='relu'),Dropout(0.4))
```

```python
#Adding first Hidden Layer
classifier.add(Dense(units=7,activation='relu'),Dropout(0.2))
```

```python
#Adding Second Hidden Layer
classifier.add(Dense(units=6,activation='relu'),Dropout(0.3))
```

```python
#Adding Second Output Layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Manually Assigning Learning Rate(Optional)
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.01)
classifier.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Early Stopping
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history = classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
y_pred  = classifier.predict(X_test)
y_pred  = (y_pred>=0.5)
```

```python
#Make the confusion matrix
from sklearn.metrics import confusion_matrix
cm  = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
##Calculate Accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,y_test)
```
